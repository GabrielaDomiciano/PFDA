The function that will do the formatting

```python
def currency_formatter(x,pos) :
    s = f"${x}"
    return s
```

```python
fig, ax = plt.subplots()
ax.bar(x,y)
ax.plot(x,y2)
ax.set_title("yo mamma")
ax.set_xlabel("x here")
ax.set_ylabel("y")
ax.xaxis.set_major_formatter (currency_formatter)

plt. show()
```

```python
fig, ax = plt.subplots()

ax.bar(x,y)
ax.plot (x,y2)

ax.set_title("yo mamma")
ax.set_xlabel("x here")
ax.set_ylabel("y")

ax.text(1, 8, "text to add", fontsize=10, verticalalignment="center")

ax.xaxis.set_major_formatter(currency_formatter)

labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment= 'right')

plt. show()
```

Annotations

```python
fig, ax = plt.subplots()

ax.bar(x,y)
ax.plot (x,y2)

ax.set_title("yo mamma")
ax.set_xlabel("x here")
ax.set_ylabel("y")

#ax.text(1, 8, "text to add", fontsize=10, verticalalignment="center")
ax.annotate('blah', (3.2, 8), xytext=(1, 9),arrowprops=dict(facecolor='black', shrink=0.05))

ax.xaxis.set_major_formatter(currency_formatter)

labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment= 'right')

plt. show()
```
